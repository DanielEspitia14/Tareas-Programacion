```python
#DATOS OBTENIDOS CON EL CODIGO DE CARLOS

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

Euler_data = pd.read_csv('Euler.csv')
RKT4_data = pd.read_csv('RKT4.csv')
analytical_data = pd.read_csv('Analytical.csv')

plt.plot(RKT4_data['t'], RKT4_data['w1'], label='$w_1$')
plt.plot(RKT4_data['t'], RKT4_data['w2'], label='$w_2$')
plt.plot(RKT4_data['t'], RKT4_data['w3'], label='$w_3$')
plt.title('Angular velocity')
plt.ylabel('$w_i$')
plt.xlabel('t')
plt.legend()
plt.show()

plt.plot(RKT4_data['t'], RKT4_data['q0'], label='$q_0$')
plt.plot(RKT4_data['t'], RKT4_data['q1'], label='$q_1$')
plt.plot(RKT4_data['t'], RKT4_data['q2'], label='$q_2$')
plt.plot(RKT4_data['t'], RKT4_data['q3'], label='$q_3$')
plt.title('Quaternion')
plt.ylabel('$q_i$')
plt.xlabel('t')
plt.legend()
plt.show()

error1 = []
error2 = []

for i, row in analytical_data.iterrows():
    v = np.array([row['w1'], row['w2'], row['w3']])
    v1 = np.array([Euler_data.iloc[i]['w1'], Euler_data.iloc[i]['w2'], Euler_data.iloc[i]['w3']])
    v2 = np.array([RKT4_data.iloc[i]['w1'], RKT4_data.iloc[i]['w2'], RKT4_data.iloc[i]['w3']])

    norm_v = np.linalg.norm(v)
    if norm_v == 0:
        norm_v = 1

    error1.append(np.linalg.norm(v1 - v) / norm_v + 1e-17)
    error2.append(np.linalg.norm(v2 - v) / norm_v + 1e-17)

plt.plot(analytical_data['t'], error1, label='Euler')
plt.plot(analytical_data['t'], error2, label='RKT4')
plt.legend()
plt.title('Angular Velocity Error')
plt.ylabel('Error')
plt.xlabel('t')
plt.yscale('log')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

#ngulos de Euler: las cantidades que rotamos el vector

phi = 0          #Precesin
theta = np.pi  #Nutacin
si = 0           #Espin


#Marices de rotacin:

#Matriz de rotacin de los ejes principales a los ejes normales
M = np.array([[np.cos(phi)*np.cos(si)- np.sin(phi)*np.cos(theta)*np.sin(si), -np.cos(phi)*np.sin(si) - np.sin(phi)*np.cos(theta)*np.cos(si),np.sin(phi)*np.sin(theta)],
  [np.sin(phi)*np.cos(si) + np.cos(phi)*np.cos(theta)*np.sin(si), -np.sin(phi)*np.sin(si) + np.cos(phi)*np.cos(theta)*np.cos(si), -np.cos(phi)*np.sin(theta)],
    [np.sin(theta)*np.sin(si), np.sin(theta)*np.cos(si), np.cos(theta)]])

#Matriz de rotacin de los ejes de Euler a los ejes principales
A = np.array([[np.sin(theta)*np.sin(si), np.cos(si), 0],
  [np.sin(theta)*np.cos(si), -np.sin(si), 0],
    [np.cos(theta), 0, 1]])


#Funciones: las necesito para rotar un vector

#Funcin para crear un cuaternin
def quaternion(a, q):       #Recibe una matriz y un arreglo de 4 dimensiones

    tr = A[0][0] + A[1][1] + A[2][2]        #Traza de la matriz

    if(tr > 0):       #Si la traza es mayor que cero, puedo expresar las componentes q1, q2, y q3 dividindolas entre q0

      q[0] = (1/2)*np.sqrt(tr + 1)
      q[1] = (A[2][1] - A[1][2])/(4*q[0])
      q[2] = (A[0][2] - A[2][0])/(4*q[0])
      q[3] = (A[1][0] - A[0][1])/(4*q[0])


    else:             #Si la traza es menor o igual a cero, no puedo dividir entre q0 y debo expresar las componentes de otra forma

      q1 = (1/2)*np.sqrt(1 + A[0][0] - A[1][1] - A[2][2])
      q2 = (1/2)*np.sqrt(1 + A[1][1] - A[2][2] - A[0][0])
      q3 = (1/2)*np.sqrt(1 + A[2][2] - A[0][0] - A[1][1])

      c = np.array([q1, q2, q3])        #Arreglo que guarda las componentes q1, q2, y q3
      max = np.max(c)                   #Retorna la mayor componente del arreglo c

      if (max == q1):                             #Si la componente q1 es el mximo, se definen el resto de componentes de la siguiente forma:
         q[0] = (A[2][1] - A[1][2])/(4*q1)
         q[1] = q1
         q[2] = (A[1][0] + A[0][1])/(4*q1)
         q[3] = (A[0][2] + A[2][0])/(4*q1)

      else:
         if (max == q2):                          #Si la componente q2 es el mximo, se definen el resto de componentes de la siguiente forma:
            q[0] = (A[0][2] - A[2][0])/(4*q2)
            q[1] = (A[0][1] + A[1][0])/(4*q2)
            q[2] = q2
            q[3] = (A[1][2] + A[2][1])/(4*q2)

         else:
            if(max == q3):                        #Si la componente q3 es el mximo, se definen el resto de componentes de la siguiente forma:
              q[0] = (A[1][0] - A[0][1])/(4*q3)
              q[1] = (A[2][0] + A[0][2])/(4*q3)
              q[2] = (A[1][2] + A[2][1])/(4*q3)
              q[3] = q3

    return(q)        #Retorna las cuatro componentes del cuaternin


#Funcin que describe el producto entre cuaterniones
def product(q1, q2):        #Recibe dos cuaterniones

  b1 = np.array([q1[1], q1[2], q1[3]])                    #Parte imaginaria del cuaternin q1
  b2 = np.array([q2[1], q2[2], q2[3]])                    #Parte imaginaria del cuaternin q2
  Re = np.array(q1[0]*q2[0] - np.dot(b1,b2))            #Parte real del producto q1q2
  Im = np.array(q1[0]*b2 + q2[0]*b1 + np.cross(b1,b2))  #Parte imaginaria del producto q1q2
  q1q2 = np.array([Re, Im[0], Im[1], Im[2]])

  return(q1q2)       #Retorna el cuaternin que sale de la multiplicacin de dos cuaterniones


#Funcin para hacer una rotacin

def rotation(q, v):       #Recibe un cuaternin y un vector de la forma (0, a, b, c)

  qc = np.array([q[0],-q[1], -q[2], -q[3]])         #Cuaternin conjugado
  qv = product(q, v)                                #Multiplicacin entre el cuaternin y el vector a rotar
  qvqc = product(qv,qc)                             #Multiplicacin entre q y v por el cuaternin conjugado

  return(qvqc)


#Ejecucin:

Q = np.zeros(4)       #Arreglo de 4 dimensiones para crear el cuaternin
q = quaternion(A,Q)   #Generador del cuaternin


#Componentes del vector a rotar en el espacio de los ngulos de Euler
phidot = 0
thetadot = np.pi
sidot = 0

v = np.array([0, phidot, thetadot, sidot])        #Vector a rotar
rv = np.array([v[1], v[2], v[3]])                 #Parte real del vector a rotar


#Rotacin:

print(rotation(q,v))

datosw = np.genfromtxt("Datos-w.dat", delimiter=",");
datosq = np.genfromtxt("Datos-q.dat", delimiter=",");

def quaternion(t):
  A=int(t/0.01)
  return datosq[A,1], datosq[A,2], datosq[A,3], datosq[A,4]

#Funcin que describe el producto entre cuaterniones
def product(q1, q2):        #Recibe dos cuaterniones

  b1 = np.array([q1[1], q1[2], q1[3]])                    #Parte imaginaria del cuaternin q1
  b2 = np.array([q2[1], q2[2], q2[3]])                    #Parte imaginaria del cuaternin q2
  Re = np.array(q1[0]*q2[0] - np.dot(b1,b2))            #Parte real del producto q1q2
  Im = np.array(q1[0]*b2 + q2[0]*b1 + np.cross(b1,b2))  #Parte imaginaria del producto q1q2
  q1q2 = np.array([Re, Im[0], Im[1], Im[2]])

  return(q1q2)       #Retorna el cuaternin que sale de la multiplicacin de dos cuaterniones


#Funcin para hacer una rotacin

def rotation(q, v):       #Recibe un cuaternin y un vector de la forma (0, a, b, c)

  qc = np.array([q[0],-q[1], -q[2], -q[3]])         #Cuaternin conjugado
  qv = product(q, v)                                #Multiplicacin entre el cuaternin y el vector a rotar
  qvqc = product(qv,qc)                             #Multiplicacin entre q y v por el cuaternin conjugado

  return np.array((qvqc))

Q=quaternion(1)
V=np.array([0,1,0,0])

rotation(Q,V)

X=[]
V=np.array([0,1,0,0])
for i in datosq[:,0]:
  Q=quaternion(i)
  X.append(rotation(Q,V))

X=np.array([np.array(x) for x in X])
plt.plot(datosq[:,0], X[:,1], c='r', label='V1')
plt.plot(datosq[:,0], X[:,2], c='g', label='V2')
plt.plot(datosq[:,0], X[:,3], c='b', label='V3')

plt.title('Rotacin del vector (1,0,0)')
plt.legend()
plt.show()

Y=[]
#V=np.array([0,(3)**(-0.5),(3)**(-0.5),(3)**(-0.5)])
V=np.array([0,0,1,0])
for i in datosq[:,0]:
  Q=quaternion(i)
  Y.append(rotation(Q,V))

Y=np.array([np.array(x) for x in Y])
plt.plot(datosq[:,0], Y[:,1], c='r', label='V1')
plt.plot(datosq[:,0], Y[:,2], c='g', label='V2')
plt.plot(datosq[:,0], Y[:,3], c='b', label='V3')

plt.title('Rotacin del vector (0,1,0)')
plt.legend()
plt.show()

Z=[]
V=np.array([0,0,0,1])
for i in datosq[:,0]:
  Q=quaternion(i)
  Z.append(rotation(Q,V))

Z=np.array([np.array(x) for x in Z])
plt.plot(datosq[:,0], Z[:,1], c='r', label='V1')
plt.plot(datosq[:,0], Z[:,2], c='g', label='V2')
plt.plot(datosq[:,0], Z[:,3], c='b', label='V3')

plt.title('Rotacin del vector (0,0,1)')
plt.legend()
plt.show()

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Supongamos que ya tienes los datos de x, y, z a lo largo del tiempo
# Reemplaza estos arreglos con tus propios datos
x_data = Y[:,1]
y_data = Y[:,2]
z_data = Y[:,3]

# Crear la figura y el objeto de la subparcela 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Inicializar el grfico con los datos iniciales
line, = ax.plot(x_data, y_data, z_data, label='Vector')

# Funcin de inicializacin
def init():
    return line,

# Funcin de actualizacin para la animacin
def update(frame):
    # Aqu puedes realizar cualquier operacin que necesites en cada fotograma
    # Actualizar el grfico con los datos existentes
    line.set_xdata(x_data)
    line.set_ydata(y_data)
    line.set_3d_properties(z_data)

    return line,

# Crear la animacin
ani = FuncAnimation(fig, update, frames=range(100), init_func=init, interval=50, blit=True)

# Mostrar la animacin
plt.show()
```
